# Mutual Community Detection Demo

This notebook demonstrates the Vertex-Pair Closeness (VPC) based community detection algorithm on the Karate Club dataset.

```python
import os
import sys
import matplotlib.pyplot as plt

# Add parent directory to path to import local module
sys.path.insert(0, os.path.abspath('..'))

from mutual_community.io import load_graph
from mutual_community.vpc import compute_vpc_weights
from mutual_community.sdp import solve_sdp
from mutual_community.rounding import hyperplane_rounding
from mutual_community.evaluation import evaluate_all
from mutual_community.visualisation import draw_communities
```

## 1. Load the Graph
We load the Zachary's Karate Club network.

```python
G = load_graph("../data/karate.gpickle")
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
```

## 2. Compute Vertex-Pair Closeness
Calculate the VPC weights for all edges to enhance the community structure.

```python
W_G, W = compute_vpc_weights(G)
print("Computed sparse adjacency matrix W.")
```

## 3. SDP Optimization & Rounding
Solve the semi-definite program (SDP) to continuous values, and then apply hyperplane rounding to retrieve discrete communities.

```python
k = 2 # Target communities
X = solve_sdp(W, k=k, alpha=1.0)

# Rounding
labels = hyperplane_rounding(X, W=W, n_trials=100)
print(f"Detected {len(set(labels))} communities.")
```

## 4. Evaluation
Assess the quality of the partition.

```python
metrics = evaluate_all(G, labels)
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```

## 5. Visualization

```python
draw_communities(G, labels, title=f"Karate Club Communities (Modularity: {metrics['Modularity']:.2f})")
```
